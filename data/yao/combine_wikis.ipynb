```python
import pandas as pd

wiki_1 = pd.read_csv('all-outputs-new-wiki-1-before-splitting.csv')
wiki_2 = pd.read_csv('all-outputs-new-wiki-2-before-splitting.csv')
```

```python
# change wiki_2's id column to id + 30
wiki_2['id'] = wiki_2['id'] + 30

wiki_total = pd.concat([wiki_1, wiki_2], ignore_index=True)
```

```python
system_list = ["Human 1 Writing", "Human 2 Writing", "GPT-3-zero-shot", "GPT-3-few-shot", "Muss", "T5-3B"]
```

```python
# Create an empty dataframe with columns: id, original, simplified, and system
df = pd.DataFrame(columns=['id', 'original', 'simplified', 'system'])

# iterate through wiki_total
for index, row in wiki_total.iterrows():
    id = row['id']
    original = row['Original Sentence']
    for system in system_list:
        simplified = row[system]
        df = df.append({'id': id, 'original': original, 'simplified': simplified, 'system': system}, ignore_index=True)
```

```python
# shuffle the dataframe with random_state=42
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# save the dataframe to csv
df.to_csv('DA_or_likert/all-outputs-new-wiki.csv', index=False)
```
